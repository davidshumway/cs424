```r
library(ggplot2)
library(rgeos)
library(ggmap)
library(maps)
library(mapdata)
library(maptools)
library(ggthemes)
library(sp)
library(stringr)
library(plyr)
library(lubridate)
```

```r
usa <- map_data("county")
il <- subset(usa, region == "illinois")
il$county = str_to_title(il$subregion)
# basic map with county boundaries and county names at the centroid of the county
getLabelPoint <- # Returns a county-named list of label points
    function(county) {Polygon(county[c('long', 'lat')])@labpt}
centroids = by(il, il$county, getLabelPoint)     # Returns list
centroids2 <- do.call("rbind.data.frame", centroids)  # Convert to Data Frame
centroids2$county = rownames(centroids)
names(centroids2) <- c('clong', 'clat', "county")
# simple map with county borders and county names
ggplot() + 
 geom_polygon(data = il, aes(x = long, y = lat, group = group), fill = "white", color = "gray") + 
 coord_fixed(1.2) + 
 geom_text(data = centroids2, aes(x = clong, y = clat, label = county), color = "darkblue", size = 2.25) +
 theme_map()
```

```r
# read in the current Covid-19 data by county in the US -
# note this command is slightly different than we used before
# with the EV data to keep things happy with this file.
covid <- read.table(file="https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv", quote = "", sep=',', header=TRUE)
# grab the Illinois part
covidIL <- subset(covid, state =="Illinois")
covidIL$fips <- NULL
# convert the dates into actual date format,
# as we have done before to make manipulating dates easier
newDates <- as.Date(covidIL$date, "%Y-%m-%d")
covidIL$newDate <- newDates
covidIL$date <- NULL
# first take a look at the data over time for Cook County
covidILCook <- subset(covidIL, county =="Cook")
ggplot(covidILCook, aes(x=newDate, y=cases)) +
 geom_point(color="blue") +
 labs(title="Cases in Cook County", x="Day", y = "Cases") + # Degrees F
 geom_line()
```

```r
#then take a look at the data for a given day on a map of Illinois counties. 
# Note that today() can be a useful function to get today's date. 
# Similarly today()-1 is yesterday. 
# Note that the data in the file may be a day or two behind.
covidILOneDay <- subset(covidIL, newDate == "2021-03-15")
ilCountyCovid <- join(il, covidILOneDay)
ggplot() +
 geom_polygon(data = ilCountyCovid, aes(x = long, y = lat, group = group, fill = cases), color = "black")+ 
 coord_fixed(1.2) +
 geom_text(data = centroids2, aes(x = clong, y = clat, label = county), color = "black", size = 2.25) +
 scale_fill_distiller(palette = "Blues") +
 labs(fill = "cases") + theme_map()

```

```r
# but since there is a huge variation between cook county and the rest of the state
# we can also convert the data to log base 10 format and show that on the map
# as is fairly common in Covid-19 visualizations.
ilCountyCovid$logCases <- log(ilCountyCovid$cases,10)
ggplot() +
 geom_polygon(data = ilCountyCovid, aes(x = long, y = lat, group = group, fill = logCases), color = "black") +
 coord_fixed(1.2) +
 geom_text(data = centroids2, aes(x = clong, y = clat, label = county), color = "black", size = 2.25) +
 scale_fill_distiller(palette = "Blues") + 
 labs(fill = "cases") + theme_map()
```
